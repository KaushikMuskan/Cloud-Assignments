```python
matrix = [1, 2, 2, 3, 3, 3]

result = []

for i in set(matrix):
    temp = []
    for j in matrix:
        if i == j:
            temp.append(j)
    result.append(temp)

print(result)
```

```python
import numpy as np

n = 3

while True:
    matrix = np.random.randint(1, 10, (n, n))
    valid = True

    for i in range(n):
        for j in range(n):
            if (matrix[i][j] + matrix[n-1-i][n-1-j]) % 2 != 0:
                valid = False

    if valid:
        break

print(matrix)
```
