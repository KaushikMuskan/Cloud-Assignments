### Aim
To generate randomized monthly sales data using NumPy and convert it into a structured Pandas DataFrame.

### Explanation
- We use NumPy to generate random sales values.
- A random seed ensures that the data is reproducible.
- The generated data represents 12 months and 4 product categories.
- Finally, the data is converted into a Pandas DataFrame for analysis.

```python
# Import required libraries
import numpy as np
import pandas as pd

# Step 1: Initializing random seed using my roll number
# This ensures reproducibility of results
np.random.seed(320059)

# Step 2: Generating random sales data
# 12 rows (months) and 4 columns
# Values between 1000 and 5000
data = np.random.randint(1000, 5000, size=(12, 4))

# Step 3: Defining month names
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Step 4: Defining product categories
columns = ["Electronics", "Clothing", "Home & Kitchen", "Sports"]

# Step 5: Converting NumPy array into Pandas DataFrame
df = pd.DataFrame(data, index=months, columns=columns)

# Display the dataset
print(df)
```

### Aim
To analyze the generated sales data using Pandas.

### Explanation
- We explore the dataset using head() and describe().
- Calculate total sales per category and per month.
- Compute growth rate using percentage change.
- Add new columns for better analysis.

```python
# Displaying first 5 rows of the dataset
print("First 5 rows:\n", df.head())

# Displaying summary statistics (mean, std, min, max, etc.)
print("\nSummary Statistics:\n", df.describe())
```

```python
# Calculating total sales for each category (column-wise sum)
category_total = df.sum()
print("Total Sales per Category:\n", category_total)

# Calculating total sales for each month (row-wise sum)
monthly_total = df.sum(axis=1)
print("\nTotal Sales per Month:\n", monthly_total)
```

```python
# Calculating percentage change between consecutive months
growth = df.pct_change()

# Calculating average growth for each category
avg_growth = growth.mean()

print("Average Growth per Category:\n", avg_growth)
```

```python
# Adding a new column 'Total Sales' (sum of all categories per month)
df["Total Sales"] = df.sum(axis=1)

# Adding 'Growth Rate' column (percentage change in total sales)
df["Growth Rate"] = df["Total Sales"].pct_change()

# Displaying updated DataFrame
print(df)
```

```python
# Applying discount based on roll number
roll_no = 320059

if roll_no % 2 == 0:
    # If roll number is even → 10% discount on Electronics
    df["Electronics"] = df["Electronics"] * 0.9
    print("10% discount applied on Electronics")
else:
    # If roll number is odd → 15% discount on Clothing
    df["Clothing"] = df["Clothing"] * 0.85
    print("15% discount applied on Clothing")

# Displaying updated DataFrame
print(df)
```

### Aim
To visualize sales data using Matplotlib and Seaborn.

### Explanation
- Line plot shows trends of sales over months.
- Box plot shows distribution and variation in sales.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ploting line graph for all categories
plt.figure()

df[["Electronics", "Clothing", "Home & Kitchen", "Sports"]].plot()

# Adding labels and title
plt.title("Monthly Sales Trends")
plt.xlabel("Months")
plt.ylabel("Sales")

# Displaying plot
plt.show()
```

### Aim
To perform statistical and mathematical operations on a NumPy array.

### Explanation
- We compute absolute values of elements.
- Find percentiles (25th, 50th, 75th) for:
  • Flattened array
  • Column-wise
  • Row-wise
- Calculate Mean, Median, and Standard Deviation for:
  • Entire array
  • Each column
  • Each row

```python
import numpy as np

array = np.array([[1, -2, 3],
                  [-4, 5, -6]])

# i. Element-wise Absolute Value
# Converts all negative values to positive
abs_array = np.abs(array)
print("Absolute Value:\n", abs_array)

# ii. Percentiles (25th, 50th, 75th)
# Flattened array (convert 2D → 1D)
flat = array.flatten()

print("\nPercentiles of Flattened Array:")
print("25th, 50th, 75th:", np.percentile(flat, [25, 50, 75]))

# Column-wise percentiles
print("\nColumn-wise Percentiles:")
print(np.percentile(array, [25, 50, 75], axis=0))

# Row-wise percentiles
print("\nRow-wise Percentiles:")
print(np.percentile(array, [25, 50, 75], axis=1))


# iii. Mean, Median, Standard Deviation
# Flattened array calculations
print("\nFlattened Array Statistics:")
print("Mean:", np.mean(flat))
print("Median:", np.median(flat))
print("Standard Deviation:", np.std(flat))

# Column-wise calculations
print("\nColumn-wise Statistics:")
print("Mean:", np.mean(array, axis=0))
print("Median:", np.median(array, axis=0))
print("Standard Deviation:", np.std(array, axis=0))

# Row-wise calculations
print("\nRow-wise Statistics:")
print("Mean:", np.mean(array, axis=1))
print("Median:", np.median(array, axis=1))
print("Standard Deviation:", np.std(array, axis=1))
```

### Aim
To perform different rounding operations on a NumPy array.

### Explanation
- Floor: Returns the largest integer less than or equal to each element.
- Ceiling: Returns the smallest integer greater than or equal to each element.
- Truncate: Removes the decimal part (towards zero).
- Round: Rounds the number to the nearest integer.

```python
import numpy as np

a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

# Floor values
# Largest integer less than or equal to each element
print("Floor:", np.floor(a))

# Ceiling values
# Smallest integer greater than or equal to each element
print("Ceiling:", np.ceil(a))

# Truncated values
# Removes decimal part (towards zero)
print("Truncate:", np.trunc(a))

# Rounded values
# Rounds to nearest integer
print("Rounded:", np.round(a))
```

### Aim
To swap two elements in a list using a temporary variable.

### Explanation
- A temporary variable is used to store one element.
- Then the values are exchanged step by step.
- This is the basic method of swapping in programming.

```python
# Given list
lst = [10, 20, 30, 40]

# Display original list
print("Original List:", lst)

# Positions to swap (example: index 0 and index 2)
i = 0
j = 2

# Step 1: Store value of first element in temporary variable
temp = lst[i]

# Step 2: Replace first element with second element
lst[i] = lst[j]

# Step 3: Replace second element with stored value
lst[j] = temp

# Display updated list
print("List after swapping:", lst)
```

### Aim
To swap two elements in a set by converting it into a list.

### Explanation
- Sets are unordered and do not support indexing.
- So, we first convert the set into a list.
- Then we swap elements using index positions.
- Finally, we convert the list back to a set.

```python
s = {10, 20, 30, 40}

# Display original set
print("Original Set:", s)

# Step 1: Convert set to list
lst = list(s)

# Step 2: Choose indices to swap (example: index 0 and index 2)
i = 0
j = 2

# Step 3: Swap using temporary variable
temp = lst[i]
lst[i] = lst[j]
lst[j] = temp

# Step 4: Convert list back to set
s = set(lst)

# Display updated set
print("Set after swapping:", s)
```
